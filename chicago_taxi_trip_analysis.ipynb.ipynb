# Prueba de Hip√≥tesis

```python
# Cargar datos del CSV
df_ohare = pd.read_csv('/datasets/project_sql_result_07.csv')

# Ver las primeras filas
display(df_ohare.head())
print(df_ohare.info())
```

```python
# Convertir fecha y hora
df_ohare['start_ts'] = pd.to_datetime(df_ohare['start_ts'])

# Crear columnas auxiliares
df_ohare['weekday'] = df_ohare['start_ts'].dt.dayofweek  # lunes=0, domingo=6
df_ohare['is_saturday'] = df_ohare['weekday'] == 5
df_ohare['is_rainy'] = df_ohare['weather_conditions'].str.lower().str.contains('rain')
df_ohare['is_rainy_saturday'] = df_ohare['is_saturday'] & df_ohare['is_rainy']
```

```python
df_ohare[['start_ts', 'weekday', 'weather_conditions', 'is_rainy_saturday']].sample(5)
```


Hip√≥tesis:
Hip√≥tesis nula (H‚ÇÄ):
La duraci√≥n promedio de los viajes en s√°bados lluviosos es igual a la de los otros d√≠as.

Hip√≥tesis alternativa (H‚ÇÅ):
La duraci√≥n promedio de los viajes en s√°bados lluviosos es diferente a la de los otros d√≠as

```python
from scipy import stats

# Crear los dos grupos
grupo_rainy_saturday = df_ohare[df_ohare['is_rainy_saturday']]['duration_seconds']
grupo_otros = df_ohare[~df_ohare['is_rainy_saturday']]['duration_seconds']

# Prueba de hip√≥tesis (asumimos varianzas desiguales)
alpha = 0.05  # Nivel de significaci√≥n

# t-test
results = stats.ttest_ind(grupo_rainy_saturday, grupo_otros, equal_var=False)

print('Valor p:', results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hip√≥tesis nula.")
    print("Conclusi√≥n: La duraci√≥n promedio **s√≠ cambia** los s√°bados lluviosos.")
else:
    print("No podemos rechazar la hip√≥tesis nula.")
    print("Conclusi√≥n: No hay evidencia suficiente para afirmar que la duraci√≥n promedio cambia.")
```

```python
print("Duraci√≥n promedio s√°bados lluviosos:", grupo_rainy_saturday.mean(), "segundos")
print("Duraci√≥n promedio otros d√≠as:", grupo_otros.mean(), "segundos")
```

## Conclusiones de la prueba de hip√≥tesis

- Se compararon dos grupos: viajes realizados los s√°bados lluviosos contra todos los dem√°s d√≠as.
- Se utiliz√≥ una prueba t de Student para muestras independientes.
- Nivel de significaci√≥n utilizado: 0.05.
- El valor p obtenido fue de X (sustituir por el real).
- Seg√∫n los resultados:
  - [ ] Rechazamos la hip√≥tesis nula.
  - [ ] No se puede rechazar la hip√≥tesis nula.

Esto indica que la duraci√≥n promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare [s√≠ / no] cambia significativamente los s√°bados lluviosos.
